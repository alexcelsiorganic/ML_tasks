```python
import pandas as pd
import numpy as np
from sklearn import cluster
```

```python
dataframe = pd.read_csv("checkins.csv", delimiter="|", skipinitialspace = True)
```

```python
dataframe.columns = dataframe.columns.str.strip()
print(dataframe.columns)
```

```python
dataframe = dataframe.dropna()
dataframe.info()
```

```python
data = dataframe.head(100000).loc[:, ['latitude', 'longitude']]
```

```python
offices = pd.read_csv("offices.txt", sep=",", skipinitialspace=True)
```

```python
clusterization = cluster.MeanShift(bandwidth=0.1)
```

```python
clusterization.fit(data)
```

```python
def distance(a, b):
    return np.sqrt(np.sum((a - b)**2))
```

```python
print(clusterization.cluster_centers_)
```

```python
labels, counts = np.unique(clusterization.labels_, return_counts=True)
print(counts)
acr = {labels[i] : clusterization.cluster_centers_[i] for i in range(len(counts))}
```

```python
def calculate_distance(dot, centers):
    m = min([distance(list(centers.values())[i], dot) for i in range(len(list(centers.values())))])
    return m
```

```python
def make_clusters(labels, centers, counts):
    d = {}
    for i in range(len(labels)):
        if (counts[i] > 15):
            d[labels[i]] = calculate_distance(centers[i], centres)
    return d
clusters = make_clusters(labels, clusterization.cluster_centers_, counts)
```

```python
print(clusters)
```

```python
clusters = {k: v for k, v in sorted(clusters.items(), key=lambda item: item[1])}
print(clusters)
```
